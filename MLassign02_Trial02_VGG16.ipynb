```python
data_path = 'dataset_input'
train = 'dataset_input/Train'
test = 'dataset_input/Test'
```

```python
categories = len(os.listdir(train))
print('Categories:', categories)
```

```python
train_info = pd.read_csv("dataset_input/Train.csv")
train_info.head()
train_info.describe()
```

```python
data = []
labels = []
channels = 3

for i in range(categories):
    path = data_path + '/Train/' + str(i)
    images = os.listdir(path)

    for img in images:
        try:
            image = cv2.imread(path + '/' + img)
            image_fromarray = Image.fromarray(image, 'RGB')
            resize_image = image_fromarray.resize((64, 64)) #tested on 32x32
            data.append(np.array(resize_image))
            labels.append(i)
        except:
            print("Error in loading image")
            
```

```python
data = np.array(data)
labels = np.array(labels)
data.shape, labels.shape
```

```python
#traffic signs
dir = data_path
    
plt.figure(figsize=(16, 16))
for i in range (0,43):
    plt.subplot(7,7,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    path = dir + "/Meta/{0}.png".format(i)
    img = plt.imread(path)
    plt.imshow(img)
    plt.xlabel(i)
```

```python
#prepare and split training, testing and validation data
```

```python
x_train, x_val, y_train, y_val = train_test_split(data, labels, test_size=0.2, random_state=42, shuffle=True)
x_train.shape, x_val.shape, y_train.shape, y_val.shape
```

```python
x_train = x_train/255 
x_val = x_val/255
```

```python
y_train = keras.utils.to_categorical(y_train, categories)
y_val = keras.utils.to_categorical(y_val, categories)
y_train.shape, y_val.shape
```

```python
vgg16 = VGG16(input_shape=(64,64,3), weights='imagenet', include_top=False)
```

```python
#use already existing weights
for layer in vgg16.layers:
    layer.trainable = False
```

```python
#define Model
add_model = Sequential()
add_model.add(Flatten(input_shape=vgg16.output_shape[1:]))
add_model.add(Dense(1024, activation='relu'))
add_model.add(Dense(y_train.shape[1], activation='softmax'))

model = Model(inputs=vgg16.input, outputs=add_model(vgg16.output))
learning_rate = 0.0001

def results(model):
  adam = Adam(lr=learning_rate)
```

```python
#compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#show model summary
#model.summary()
```

```python
#train model
epochs = 3 #Changed from 10 to 3 cause it took longer to process bigger images
history = model.fit(x_train, y_train, batch_size=128, epochs=epochs, validation_data=(x_val, y_val))
```

```python
_,accuracy=model.evaluate(x_val,y_val)
```
